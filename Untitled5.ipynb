```python
def SecondSpace():
    Erjian_data = {
        '定位': '微握拳，當手食指本節（第2掌指關節）前橈側凹陷中。',
        '解剖': '有指屈淺、深肌腱；有來自橈動脈的指背及掌側動、靜脈，布有橈神經的指背側固有神經，正中神經的指掌側固有神經。',
        '主治': '目昏，鼻衄，齒痛，口歪，咽喉腫痛,熱病。',
        '配伍': '配合谷治齒痛。',
        '刺灸法': '直刺0.2～0.3寸。'
    }
    return Erjian_data

def PoolattheCrook():
    Quchi_data = {
        '定位': '在肘橫紋外側端，屈肘，當尺澤與肱骨外上髁連線中點。',
        '解剖': '橈側腕長伸肌起始部，肱橈肌的撓側；有橈返動脈的分支；布有前臂背側皮神經,內側深層為橈神經本干。',
        '主治': '咽喉腫痛，齒痛，目赤痛，瘰癧，癮疹，熱病上肢不遂，手臂腫痛，腹痛吐瀉，高血壓，癲狂。',
        '配伍': '配血海,足三里治癮疹；配手三里治上肢不遂；配太沖、大椎治高血壓。',
        '刺灸法': '直刺 1～1.5寸。'
    }
    return Quchi_data

def JoiningValley():
    HeGu_data = {
        '定位': '在手背，第1、2掌骨間，當第2掌骨橈側的中點處。',
        '解剖': '在第一、二掌骨之間，第一骨間背側肌中，深層有拇收肌橫頭；有手背靜脈網,為頭靜脈的起部，腧穴近側正當橈動脈從手背穿向手掌之處；布有橈神經淺支的掌背側神經，深部有正中神經的指掌側固有神經。',
        '主治': '頭痛，目赤腫痛，鼻衄，齒痛，牙關緊閉，口眼歪斜，耳聾，痄腮，咽喉腫痛,熱病無汗，多汗，腹痛，便秘，經閉，滯產。',
        '配伍': '配太陽治頭痛;配太沖治目赤腫痛;配迎香治鼻疾；配少商治咽喉腫痛; 配三陰交治經閉,滯產; 配地倉頰車治眼歪斜。',
        '刺灸法': '直刺0.5～1寸。'
    }
    return HeGu_data

def WelcomeFragrance():
    Yingxiang_data = {
        '定位': '在鼻翼外緣中點旁，當鼻唇溝中間。',
        '解剖': '在上唇方肌中，深部為梨狀孔的邊緣; 有面動 、靜脈及眶下動 、靜脈分支 ；布有面神經與眶下神經的吻合叢。',
        '主治': '鼻塞，鼽衄，口歪，面癢，膽道蛔蟲症。',
        '配伍': '無。',
        '刺灸法': '斜刺或平刺0.3～0.5寸。'
    }
    return Yingxiang_data

def ElbowCrevice():
    Zhouliao_data = {
        '定位': '在臂外側，屈肘，曲池上方 1寸，當肱骨邊緣處。',
        '解剖': '在橈骨外上髁上緣肱肌起始部，肱三頭肌外緣；有橈側副動脈；布有前臂背側皮神經及橈神經。',
        '主治': '肘臂部瘓痛，麻木，攣急。',
        '配伍': '配曲池治肘臂疾病。',
        '刺灸法': '直刺0.5～1寸。'
    }
    return Zhouliao_data

def large_intestine_channel_of_Hand_Yangming():
    acupoints_info = {
        '二間穴': SecondSpace(),
        '曲池穴': PoolattheCrook(),
        '合谷穴': JoiningValley(),
        '迎香穴': WelcomeFragrance(),
        '肘髎穴': ElbowCrevice()
        # ... 其他穴位
    }
    return acupoints_info

def get_acupoint_info(acupoint_name, info_category=None):
    acupoints_info = large_intestine_channel_of_Hand_Yangming()

    acupoint_data = acupoints_info.get(acupoint_name, None)
    if acupoint_data is None:
        return "未找到指定的穴位信息。"

    if info_category:
        return acupoint_data.get(info_category, "未找到指定類別的信息。")

    return acupoint_data

def query_acupoint():
    while True:
        print("\n--- 查詢系統 ---")
        acupoint_name = input("請輸入您要查詢的穴位名稱（輸入'退出'結束查詢）:\n(二間穴、曲池穴、合谷穴、迎香穴、肘髎穴選擇其一owo):")
        if acupoint_name == "退出":
            print("\n--- 已退出查詢 ---")
            print("感謝使用本系統，祝你好運owo")
            break

        acupoint_info = get_acupoint_info(acupoint_name)
        if "未找到" in acupoint_info:
            print(f"\n{acupoint_info}\n")
            continue

        while True:
            print(f"\n找到穴位：{acupoint_name}")
            print("可查詢的信息類別有：定位、解剖、主治、配伍、刺灸法")
            info_category = input("請輸入您要查詢的信息類別（輸入'全部'顯示所有信息，或輸入'返回'返回上一層）:")

            if info_category == "返回":
                print("\n返回上一層...\n")
                break
            elif info_category == "全部":
                print(f"\n{acupoint_name} - 全部信息:")
                for category, detail in acupoint_info.items():
                    print(f"{category}: {detail}")
                print()  # 為了排版整潔owo
            elif info_category in ["定位", "解剖", "主治", "配伍", "刺灸法"]:
                category_info = get_acupoint_info(acupoint_name, info_category)
                print(f"\n{acupoint_name} - {info_category}:\n{category_info}\n")
            else:
                print("\n輸入的信息類別不正確，請重新輸入。\n")

            # 再次查詢或返回
            while True:
                next_action = input("輸入'再查詢'以查詢此穴位的其他信息，或輸入'返回'返回穴位選擇：")
                if next_action in ["再查詢", "返回"]:
                    break
                else:
                    print("請輸入有效的指令（'再查詢' 或 '返回'）。")
            if next_action == "返回":
                print("\n返回穴位選擇...")
                break
# 啟動查詢
query_acupoint()
```
